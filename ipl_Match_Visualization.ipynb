```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load dataset
ipl_df=pd.read_csv('/content/ipl (2).csv')
```

#  Matches Played Per Season

```python
matches_per_season = ipl_df.groupby('season')['id'].count()

plt.figure(figsize=(10,5))
plt.plot(matches_per_season.index, matches_per_season.values, marker='o')
plt.title('Matches Played Per Season')
plt.xlabel('Season')
plt.ylabel('Number of Matches')
plt.grid(True)
plt.show()
```

# Average Result Margin Per Season

```python
avg_margin = ipl_df.groupby('season')['result_margin'].mean()

plt.figure(figsize=(10,5))
plt.plot(avg_margin.index, avg_margin.values,color='black',marker='>',markerfacecolor='red',label='Avg Margin')
plt.title('Average Result Margin Per Season')
plt.xlabel('Season')
plt.ylabel('Average Result Margin')
plt.grid(True)
plt.show()
```

# Super Over Matches Per Season


```python
super_overs=ipl_df[ipl_df['super_over']==True]
season_wise=super_overs.groupby('season').size()

plt.figure(figsize=(10,5))
season_wise.plot(color='darkblue',marker='o',markerfacecolor='red')
plt.title('Super Over Matches Per Season')
plt.xlabel('Season')
plt.ylabel('Count of Super Overs')
plt.grid(True)
plt.show()
```

# Toss Decision Trend Over Seasons

```python
bat_decision = (
    ipl_df[ipl_df['toss_decision'] == 'bat']
    .groupby('season')['id']
    .count()
)

field_decision = (
    ipl_df[ipl_df['toss_decision'] == 'field']
    .groupby('season')['id']
    .count()
)

plt.figure(figsize=(10,5))

plt.plot(bat_decision.index, bat_decision.values,
         marker='o', label='Bat First')

plt.plot(field_decision.index, field_decision.values,
         marker='o', label='Field First')

plt.title('Toss Decision Trend Over Seasons')
plt.xlabel('Season')
plt.ylabel('Count')
plt.legend()
plt.grid(True)
plt.show()
```

# RCB Season wise Wins

```python
rcb=ipl_df[ipl_df['winner']=='Royal Challengers Bengaluru']
rcb_count=rcb.groupby('season')['winner'].count()
```

```python
plt.plot(rcb_count,color='red',linewidth=4,linestyle='solid',marker='o',markerfacecolor='white',markeredgecolor='black',markersize=5,label="RCB")
plt.legend(loc='upper right',fontsize=10,shadow=True,facecolor='white',edgecolor='red',title='Team',)
plt.grid()
plt.xlabel('Season',color='red',fontweight='bold',fontsize=15)
plt.ylabel('Winning Count',color='red',fontweight='bold',fontsize=15)
plt.xticks(rcb_count.index,rotation=45)
plt.yticks(rcb_count.values,rotation=45)
plt.annotate('Max Wine',(rcb_count.idxmax(),rcb_count.max()),color='red',xytext=(2022,11),arrowprops={})
plt.annotate('Min Wine',(rcb_count.idxmin(),rcb_count.min()),color='red',xytext=(2022,1),arrowprops={})
plt.show()
```

# TEAMS

```python
rcb=ipl_df[ipl_df['winner']=='Royal Challengers Bengaluru']
rcb_count=rcb.groupby('season')['winner'].count()

mi=ipl_df[ipl_df['winner']=='Mumbai Indians']
mi_count=mi.groupby('season')['winner'].count()

csk=ipl_df[ipl_df['winner']=='Chennai Super Kings']
csk_count=csk.groupby('season')['winner'].count()

rr=ipl_df[ipl_df['winner']=='Rajasthan Royals']
rr_count=rr.groupby('season')['winner'].count()

kkr=ipl_df[ipl_df['winner']=='Kolkata Knight Riders']
kkr_count=kkr.groupby('season')['winner'].count()

srh=ipl_df[ipl_df['winner']=='Sunrisers Hyderabad']
srh_count=srh.groupby('season')['winner'].count()

```

```python
fig , axis=plt.subplots(2,3,sharex=True,sharey=True,figsize=(20,15),edgecolor='gray',linewidth=8)

fig.set_facecolor('#03AED2')
fig.suptitle("IPL Analysis Dashboard horizental line show max in year (2008-2024)".upper(),
             fontsize=25,color='blue')

# RCB

axis[0][0].plot(rcb_count,color='blue',linewidth=4,linestyle='solid',marker='o',markerfacecolor='white',markeredgecolor='black',markersize=5,label="RCB")
axis[0][0].set_yticks(np.arange(0,15))
axis[0][0].set_facecolor('red')
axis[0][0].axhline(y=rcb_count.max(),color='white',lw=5)
axis[0][0].axvline(x=rcb_count.idxmax(),color='white')


# CSK

axis[0][1].plot(csk_count,color='blue',linewidth=4,linestyle='solid',marker='o',markerfacecolor='white',markeredgecolor='black',markersize=5,label="mi")
axis[0][1].set_yticks(np.arange(0,15))
axis[0][1].set_facecolor('yellow')
axis[0][1].axhline(y=csk_count.max(),color='white',lw=5)
axis[0][1].axvline(x=csk_count.idxmax(),color='white')

# MI

axis[0][2].plot(mi_count,color='white',lw=5,marker='o',markersize=10)
axis[0][2].set_facecolor('blue')
axis[0][2].axhline(y=mi_count.max(),color='white',lw=5)
axis[0][2].axvline(x=mi_count.idxmax(),color='white')

# RR

axis[1][0].plot(rr_count,color='blue',lw=5,marker='o',markersize=10)
axis[1][0].set_facecolor('pink')
axis[0][2].axhline(y=rr_count.max(),color='white',lw=5)
axis[0][2].axvline(x=rr_count.idxmax(),color='white')

# KKR

axis[1][1].plot(kkr_count,color='blue',lw=5,marker='o',markersize=10)
axis[1][1].set_facecolor('purple')
axis[1][1].axhline(y=kkr_count.max(),color='white',lw=5)
axis[1][1].axvline(x=kkr_count.idxmax(),color='white')

# SRH

axis[1][2].plot(srh_count,color='blue',lw=5,marker='o',markersize=10)
axis[1][2].set_facecolor('#FF7F11')
axis[1][2].axhline(y=srh_count.max(),color='white',lw=5)
axis[1][2].axvline(x=srh_count.idxmax(),color='white')


titles=iter(['RCB','CSK','MI','RR','KKR','SRH'])
for ax in axis:
  for i in range(0,3):
    ax[i].set_title(next(titles),fontsize=20,fontweight='bold',color='#546B41')

for ax in axis:
  for i in range(0,3):
    ax[i].grid()

axis[0][0].tick_params(axis='x',rotation=90)

# fig.savefig('dash.jpg')

fig.tight_layout()
fig.show()
```
