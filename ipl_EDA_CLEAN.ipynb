```python
ipl['city']=ipl['city'].fillna('Unknown')
```

```python
pd.concat([ipl['team1'],ipl['team2']]).unique()
```

```python
ipl['team1']=ipl['team1'].replace({'Royal Challengers Bangalore':'Royal Challengers Bengaluru',
                     'Kings XI Punjab':'Punjab Kings',
                     'Delhi Daredevils':'Delhi Capitals',
                     'Deccan Chargers':'Sunrisers Hyderabad',
                     'Pune Warriors':'Rising Pune Supergiants',
                     'Rising Pune Supergiants':'Rising Pune Supergiant'})
```

```python
ipl['team2']=ipl['team2'].replace({'Royal Challengers Bangalore':'Royal Challengers Bengaluru',
                     'Kings XI Punjab':'Punjab Kings',
                     'Delhi Daredevils':'Delhi Capitals',
                     'Deccan Chargers':'Sunrisers Hyderabad',
                     'Pune Warriors':'Rising Pune Supergiants',
                     'Rising Pune Supergiants':'Rising Pune Supergiant'})
```

```python
ipl['season']=ipl['season'].astype(int)
```

# we are Filling player of match is No award when match was suspended

# we are filling winner is no result when match was suspended

**10.WHAT IS THE WIN PERCENTAGE OF EACH TEAMS**

```python
teams_played=pd.concat([ipl['team1'],ipl['team2']]).value_counts()
teams_wins=ipl['winner'].value_counts()
```

```python
win_ratio={}
for team in pd.concat([ipl['team1'],ipl['team2']]).unique():
  win_ratio[team]= teams_wins[team]/teams_played[team]*100
```

```python
for key , value in win_ratio.items():
  print(f'{key} Team Win Ratio Is {value}',end='\n')
```
